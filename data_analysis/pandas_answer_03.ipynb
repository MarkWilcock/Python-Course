## Setup

Run this cell to import pandas and create the sample dataset.

```python
import pandas as pd
import numpy as np

# Sales data for a company with multiple stores
sales_data = {
    'store_id': ['S001', 'S001', 'S001', 'S001', 'S002', 'S002', 'S002', 'S002', 
                 'S003', 'S003', 'S003', 'S003', 'S004', 'S004', 'S004', 'S004'],
    'region': ['North', 'North', 'North', 'North', 'North', 'North', 'North', 'North',
               'South', 'South', 'South', 'South', 'South', 'South', 'South', 'South'],
    'product': ['Laptop', 'Phone', 'Tablet', 'Headphones', 
                'Laptop', 'Phone', 'Tablet', 'Headphones',
                'Laptop', 'Phone', 'Tablet', 'Headphones',
                'Laptop', 'Phone', 'Tablet', 'Headphones'],
    'quantity': [15, 45, 30, 60, 20, 50, 25, 55, 18, 40, 28, 50, 12, 35, 22, 45],
    'price': [1200, 800, 500, 150, 1200, 800, 500, 150, 1200, 800, 500, 150, 1200, 800, 500, 150],
    'discount_pct': [10, 5, 8, 0, 12, 8, 10, 5, 8, 5, 6, 0, 15, 10, 12, 8]
}

sales = pd.DataFrame(sales_data)

# Calculate revenue (quantity * price)
sales['revenue'] = sales['quantity'] * sales['price']

print("Sales DataFrame created successfully!")
print(f"Shape: {sales.shape}")
sales.head(10)
```

## Part 1: Basic Grouping and Aggregation

### Exercise 1.1: Total Revenue by Store

```python
# Solution
total_revenue_by_store = sales.groupby('store_id')['revenue'].sum()
print("Total revenue by store:")
print(total_revenue_by_store)
```

### Exercise 1.2: Average Quantity by Product

```python
# Solution
avg_quantity_by_product = sales.groupby('product')['quantity'].mean()
print("Average quantity by product:")
print(avg_quantity_by_product)
```

### Exercise 1.3: Count Sales by Region

```python
# Solution 1: Using size()
sales_count_by_region = sales.groupby('region').size()
print("Sales count by region (using size):")
print(sales_count_by_region)

# Solution 2: Using count()
print("\nSales count by region (using count):")
print(sales.groupby('region')['store_id'].count())
```

## Part 2: Multiple Aggregation Functions

### Exercise 2.1: Multiple Statistics for Revenue

```python
# Solution
revenue_stats_by_region = sales.groupby('region')['revenue'].agg(['sum', 'mean', 'min', 'max'])
print("Revenue statistics by region:")
print(revenue_stats_by_region)
```

### Exercise 2.2: Different Aggregations for Different Columns

```python
# Solution
multi_agg_by_store = sales.groupby('store_id').agg({
    'quantity': ['sum', 'mean'],
    'revenue': ['sum', 'max'],
    'discount_pct': 'mean'
})
print("Multiple aggregations by store:")
print(multi_agg_by_store)
```

## Part 3: Grouping by Multiple Columns

### Exercise 3.1: Revenue by Region and Product

```python
# Solution
revenue_by_region_product = sales.groupby(['region', 'product'])['revenue'].sum()
print("Total revenue by region and product:")
print(revenue_by_region_product)
```

### Exercise 3.2: Average Discount by Region and Product

```python
# Solution
avg_discount = sales.groupby(['region', 'product'])['discount_pct'].mean().round(2)
print("Average discount percentage by region and product:")
print(avg_discount)
```

## Part 4: Understanding size() vs count()

### Exercise 4.1: Compare size() and count()

```python
# Solution
# 1. Using size()
size_by_region = sales.groupby('region').size()
print("Using size() - counts all rows per region:")
print(size_by_region)

# 2. Using count()
count_by_region = sales.groupby('region').count()
print("\nUsing count() - counts non-null values per column per region:")
print(count_by_region)

# 3. Comparison
print("\nComparison:")
print("size() returns a Series with one value per group (number of rows)")
print("count() returns a DataFrame with counts for each column")
print("Since there are no missing values, all count values equal the size values.")
```

## Part 5: Transform vs Aggregate

### Exercise 5.1: Add Region Average to Each Row

```python
# Solution
sales['region_avg_revenue'] = sales.groupby('region')['revenue'].transform('mean')

# Display a few rows to verify
print("Region average revenue added to each row:")
print(sales[['store_id', 'region', 'revenue', 'region_avg_revenue']].head(10))
```

### Exercise 5.2: Calculate Performance vs Store Average

```python
# Solution
# 1. Add store average quantity to each row
sales['store_avg_quantity'] = sales.groupby('store_id')['quantity'].transform('mean')

# 2. Calculate difference from store average
sales['diff_from_store_avg'] = sales['quantity'] - sales['store_avg_quantity']

# Display results
print("Performance vs store average:")
print(sales[['store_id', 'product', 'quantity', 'store_avg_quantity', 'diff_from_store_avg']].head(8))
```

## Part 6: Filtering Groups

### Exercise 6.1: Filter High-Performing Stores

```python
# Solution
high_performing_stores = sales.groupby('store_id').filter(lambda x: x['revenue'].sum() > 100000)
print("Stores with total revenue > 100,000:")
print(high_performing_stores[['store_id', 'product', 'revenue']])
print(f"\nUnique stores remaining: {high_performing_stores['store_id'].unique()}")
```

### Exercise 6.2: Filter Products with High Average Quantity

```python
# Solution
high_volume_products = sales.groupby('product').filter(lambda x: x['quantity'].mean() >= 40)
print("Products with average quantity >= 40:")
print(high_volume_products[['product', 'quantity']])
print(f"\nUnique products remaining: {high_volume_products['product'].unique()}")
```

## Part 7: Custom Aggregation Functions

### Exercise 7.1: Calculate Range

```python
# Solution
def calc_range(x):
    return x.max() - x.min()

# Use the function with groupby
quantity_range_by_store = sales.groupby('store_id')['quantity'].agg(calc_range)
print("Range of quantity by store:")
print(quantity_range_by_store)

# Alternative: using both built-in and custom functions
print("\nWith multiple aggregations:")
print(sales.groupby('store_id')['quantity'].agg(['min', 'max', calc_range]))
```

## Part 8: Challenge Exercises

### Challenge 8.1: Top Product per Region

```python
# Solution 1: Using idxmax
revenue_by_region_product = sales.groupby(['region', 'product'])['revenue'].sum()
top_product_per_region = revenue_by_region_product.groupby('region').idxmax()
print("Top product per region (using idxmax):")
print(top_product_per_region)

# Solution 2: Using sort_values and groupby with first()
print("\nAlternative approach with sort_values:")
region_product_revenue = sales.groupby(['region', 'product'], as_index=False)['revenue'].sum()
top_products = region_product_revenue.sort_values('revenue', ascending=False).groupby('region').first()
print(top_products)
```

### Challenge 8.2: Performance Ranking

```python
# Solution
# First, let's see total revenue per store
store_revenue = sales.groupby(['region', 'store_id'])['revenue'].sum().reset_index()
print("Store revenue:")
print(store_revenue)

# Rank stores within each region (1 = highest revenue)
store_revenue['store_rank'] = store_revenue.groupby('region')['revenue'].rank(ascending=False, method='dense')
print("\nStore rankings by region:")
print(store_revenue.sort_values(['region', 'store_rank']))

# Add rank back to original sales DataFrame
sales = sales.merge(store_revenue[['store_id', 'store_rank']], on='store_id', how='left')
print("\nRank added to sales DataFrame:")
print(sales[['store_id', 'region', 'revenue', 'store_rank']].head(8))
```

### Challenge 8.3: Summary Report

```python
# Solution
summary_report = sales.groupby('region').agg({
    'revenue': 'sum',
    'store_id': 'nunique',  # nunique counts unique values (number of stores)
    'quantity': 'sum',
    'discount_pct': 'mean'
})

# Rename columns for clarity
summary_report.columns = ['Total Revenue', 'Number of Stores', 'Total Quantity', 'Avg Discount %']

# Calculate average revenue per store
summary_report['Avg Revenue per Store'] = summary_report['Total Revenue'] / summary_report['Number of Stores']

# Round for better display
summary_report = summary_report.round(2)

print("Summary Report by Region:")
print(summary_report)
```

## Bonus: Reflection Questions - ANSWERS

**1. When would you use `.transform()` instead of `.agg()`?**

Use `.transform()` when you need to add group-level statistics back to each row in the original DataFrame, maintaining the same shape. For example:
- Adding each region's average to every row in that region
- Calculating differences from group means
- Creating percentage of group total columns

Use `.agg()` when you want summary statistics with one row per group.

**2. What's the difference between `.size()` and `.count()`?**

- `.size()`: Returns the total number of **rows** in each group, including rows with null values. Returns a Series.
- `.count()`: Returns the number of **non-null values** in each column for each group. Returns a DataFrame with counts for each column.

If there are no missing values, they give the same counts. If there are missing values, `.size()` includes them but `.count()` excludes them.

**3. Give an example of when you might need to group by multiple columns in a real-world scenario.**

Examples:
- **E-commerce:** Group by `category` and `brand` to analyze sales by product type and manufacturer
- **Finance:** Group by `year` and `quarter` to analyze revenue trends over time
- **Healthcare:** Group by `hospital` and `department` to analyze patient volumes
- **Education:** Group by `school` and `grade_level` to analyze test scores
- **Marketing:** Group by `campaign` and `channel` to compare advertising performance

**4. What advantage does using a custom aggregation function give you?**

Custom aggregation functions allow you to:
- Perform calculations not available in built-in functions (e.g., range, weighted averages, percentiles)
- Apply domain-specific business logic (e.g., "count orders above threshold")
- Combine multiple operations in one function (e.g., "coefficient of variation" = std/mean)
- Make your code more readable by giving meaningful names to complex calculations
- Reuse the same calculation across different analyses

## Additional Practice Examples

### Bonus Example 1: Named Aggregations

```python
# Using named aggregations for clearer column names (pandas 0.25+)
region_summary = sales.groupby('region').agg(
    total_revenue=('revenue', 'sum'),
    avg_revenue=('revenue', 'mean'),
    total_items=('quantity', 'sum'),
    num_transactions=('store_id', 'count')
).round(2)

print("Region summary with named aggregations:")
print(region_summary)
```

### Bonus Example 2: Percentage of Total

```python
# Calculate store revenue as percentage of region total
store_revenue_totals = sales.groupby('store_id')['revenue'].sum()
region_revenue_totals = sales.groupby('region')['revenue'].sum()

# Create summary
store_summary = pd.DataFrame({
    'store_revenue': store_revenue_totals,
    'region': sales.groupby('store_id')['region'].first()
})

# Add region total and calculate percentage
store_summary['region_total'] = store_summary['region'].map(region_revenue_totals)
store_summary['pct_of_region'] = (store_summary['store_revenue'] / store_summary['region_total'] * 100).round(2)

print("Store contribution to region revenue:")
print(store_summary)
```

### Bonus Example 3: Aggregating with Conditions

```python
# Count high-discount sales per store
high_discount_count = sales.groupby('store_id').apply(lambda x: (x['discount_pct'] > 8).sum())
print("Number of sales with discount > 8% per store:")
print(high_discount_count)

# Alternative using custom function with agg
def count_high_discount(x):
    return (x > 8).sum()

print("\nUsing custom function:")
print(sales.groupby('store_id')['discount_pct'].agg(count_high_discount))
```
